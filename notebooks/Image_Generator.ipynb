```python
from tensorflow.keras.models import load_model
from tensorflow.keras.optimizers import Adam
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import os
```

```python
from google.colab import drive

drive.mount('/content/drive/')

DIR_ROOT = '/content/drive/MyDrive/Colab Notebooks/Progetto DeepLearning'
```

```python
os.makedirs(DIR_ROOT+'/GANgeneratedImages')

TO_SAVE_PATH = DIR_ROOT+'/GANgeneratedImages'


os.makedirs(TO_SAVE_PATH+'/covid')
os.makedirs(TO_SAVE_PATH+'/normal')
os.makedirs(TO_SAVE_PATH+'/viral')
```

```python
covid_to_Gen = 6000
normal_to_gen = 100
viralPneu_to_gen = 8000


#model_to_load = DIR_ROOT+'/saved_model/generator_epoch_1499.hdf5'

model_to_load = DIR_ROOT+'/saved_model/generator_epoch_1493.hdf5'
```

```python
model = load_model(model_to_load)
```

```python
optimizer = Adam(0.0002, 0.5)
losses = ['binary_crossentropy', 'sparse_categorical_crossentropy']

model.compile(loss=losses,optimizer=optimizer,metrics=['accuracy'])
```

###Generate Covid Images

```python
for i in range(covid_to_Gen):
       noise = np.random.normal(0, 0.02, (1, 1000))
       sampled_labels = np.array(0)
       X = model.predict([sampled_labels.reshape((-1, 1)),noise], verbose=0)
       X = (X * 127.5 + 127.5).astype(np.uint8)

       image = Image.fromarray(X[0,:,:,0], "L")
       image.save(TO_SAVE_PATH+'/covid/' +"COVID-" + str(i + 1) + "S.jpeg")
```

###Generate Normal Images

```python
for i in range(normal_to_gen):
       noise = np.random.normal(0, 0.02, (1, 1000))
       sampled_labels = np.array(0)
       X = model.predict([sampled_labels.reshape((-1, 1)),noise], verbose=0)
       X = (X * 127.5 + 127.5).astype(np.uint8)

       image = Image.fromarray(X[0,:,:,0], "L")
       image.save(TO_SAVE_PATH+'/normal/' +"Normal-" + str(i + 1) + "S.jpeg")
```

```python
for i in range(viralPneu_to_gen):
       noise = np.random.normal(0, 0.02, (1, 1000))
       sampled_labels = np.array(0)
       X = model.predict([sampled_labels.reshape((-1, 1)),noise], verbose=0)
       X = (X * 127.5 + 127.5).astype(np.uint8)

       image = Image.fromarray(X[0,:,:,0], "L")
       image.save(TO_SAVE_PATH+'/viral/' +"Viral Pneumonia-" +str(i + 1) + "S.jpeg")
```
